## Importing Modules

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report, average_precision_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import roc_auc_score, roc_curve, auc, PrecisionRecallDisplay, precision_recall_curve
from xgboost import XGBClassifier

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

print('Modules Imported')
```

## Loading the Dataset

```python
file_path = 'term-deposit-marketing-2020.csv'

data = pd.read_csv(file_path)
```

## Exploring the Dataset

**X:**
- `age`: age of customer (numeric)

- `job`: type of job (categorical)

- `marital`: marital status (categorical)

- `education`: (categorical)

- `default`: has credit in default? (binary)

- `balance`: average yearly balance, in euros (numeric)

- `housing`: has a housing loan? (binary)

- `loan`: has personal loan? (binary)

- `contact`: contact communication type (categorical)

- `day`: last contact day of the month (numeric)

- `month`: last contact month of year (categorical)

- `duration`: last contact duration, in seconds (numeric)

- `campaign`: number of contacts performed during this campaign and for this client (numeric, includes last contact)

**Y:**
- `subscribed`: has the client subscribed to a term deposit? (binary)

Renaming columns in favor of more descriptive names.

## Data Analysis

### Class Balance

```python
customers = data.groupby('subscribed').size()

total_customers = data['subscribed'].count()

customers_opinions = ['Not Subscribed', 'Subscribed']

color = ['r','g']

types_of_customers = data.groupby('subscribed').size().index

plt.pie(customers*100/total_customers, labels = customers_opinions, colors= color, radius=0.85, autopct='%1.1f%%', wedgeprops={'alpha': 0.7})
plt.legend(loc='lower right')
plt.title("Distribution of Clients")
plt.show()
```

**Severe Class Imbalance:** The pie chart reveals a highly imbalanced dataset, with **92.8%** of customers (37,104 out of 40,000 total records) **not subscribing** to the term deposit, compared to only **7.2%** (2,896 records) who **did subscribe**. This imbalance highlights that non-subscriptions dominate the data, which could bias machine learning models toward predicting "no subscription" by default.

### Age

```python
age_data = data.copy()

# Setting the intervals for the data
labels = ['18-25','26-33','34-41','42-49','50-59', '+60']
cut_bins = [18, 25, 33, 41, 49, 59, float("inf")]

age_data['age_intervals'] = pd.cut(data['age'], bins = cut_bins, labels = labels)

subs = (
    age_data[age_data['subscribed']=='yes']
    .groupby('age_intervals', observed=False)
    .size()
)

non_subs = (
    age_data[age_data['subscribed']=='no']
    .groupby('age_intervals', observed=False)
    .size()
)

percentage = (
    subs
    .div(subs + non_subs)
    .fillna(0)
    * 100
)

# Setting the data for the satisfied clients
n_satisfied_clients = age_data[age_data['subscribed'] == 'yes']['age'].value_counts().sort_index()

# Setting the data for the unsatisfied clients
n_unsatisfied_clients = age_data[age_data['subscribed'] == 'no']['age'].value_counts().sort_index()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

# First Graph
sns.histplot(data=age_data, x='age', hue='subscribed', 
             multiple='stack',          
             bins=30, 
             ax=ax1,
             palette=['red', 'green'],  
             edgecolor='.3',
             linewidth=.5)

ax1.set_ylabel("Clients")
ax1.set_xlabel("Age")
ax1.set_title("Clients by Age",fontsize=14)

# Second Graph
ax2.bar(labels, percentage)
ax2.set_ylabel("Percentage of Subscribers (%)")
ax2.set_xlabel("Age Range")
ax2.set_title("Probability Subscribed by Age",fontsize=14)
```

The analysis indicates that **younger clients (18–25)** and **older clients (60+)** present a higher proportion of subscriptions compared to other age groups. While these groups represent a smaller portion of the total client base, the **probability of subscription within these ranges is noticeably higher**.

In contrast, the majority of clients fall within the **26–59 age range**, where the overall number of contacts is significantly larger, but the subscription rate is comparatively lower.

These findings suggest that age has a meaningful relationship with subscription behavior. From a strategic perspective, it may be beneficial to **increase outreach efforts toward the 18–25 and 60+ age groups**, as these segments demonstrate a higher likelihood of converting into subscribers.

Although these groups are smaller in absolute numbers, focusing more resources on them could potentially **improve the overall conversion rate of the marketing campaign**.

### Job

```python
# Setting the data for the satisfied clients
n_satisfied_clients = (
    data[data['subscribed'] == 'yes']
    .groupby('job')
    .size()
)

# Setting the data for the unsatisfied clients
n_unsatisfied_clients = (
    data[data['subscribed'] == 'no']
    .groupby('job')
    .size()
)

# Setting the index for all clients
all_clients = sorted(data['job'].unique())

# Filling gaps of non-existing data
n_satisfied_clients = n_satisfied_clients.reindex(all_clients, fill_value=0)
n_unsatisfied_clients = n_unsatisfied_clients.reindex(all_clients, fill_value=0)

total_customers = n_satisfied_clients + n_unsatisfied_clients

percentage = (
    n_satisfied_clients
    .div(total_customers)
    .fillna(0)
    * 100
)

# Standardizing the x axis for both graphs
x_axis = data.groupby('job').size().index

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(20,16))

# First Graph
ax1.bar(x_axis, n_satisfied_clients, label='Yes', color='g')
ax1.bar(x_axis, n_unsatisfied_clients, bottom=n_satisfied_clients, label='No', color='r')

ax1.set_ylabel("Clients")
ax1.set_xlabel("Job")
ax1.set_title("Clients by Job",fontsize=14)
ax1.legend(loc='upper right', title='subscribed')

# Second Graph
ax2.bar(x_axis, percentage)
ax2.set_ylabel("Percentage of Subscribers (%)")
ax2.set_xlabel("Job")
ax2.set_title("Probability Subscribed by Job",fontsize=14)
```

The analysis of subscription rates across different job categories suggests that **occupation does not strongly influence the likelihood of a client subscribing**. Most job groups present relatively similar subscription proportions, indicating that job type alone is not a strong predictor of subscription behavior.

One apparent exception is the **student** category, which shows a higher subscription percentage. However, this result should be interpreted with caution because the number of clients in this group is relatively small. Smaller sample sizes can produce higher variability and may exaggerate the apparent effect.

Overall, the results indicate that **job category has limited impact on subscription outcomes** compared to other variables in the dataset. As a result, targeting marketing efforts based solely on occupation may not significantly improve campaign performance.

### Balance

```python
balance_data = data.copy()

# Setting the intervals for the data
labels = ['-1000', '-999-0', '1-499' ,'500-799', '800-999', '1000-1299','1300-1499', '1500-1999', '2000-4999', '+5000']
cut_bins = [float("-inf"), -999, 0, 499, 799, 999, 1299, 1499, 1999, 4999, float("inf")]

balance_data['balance_intervals'] = pd.cut(data['balance'], bins = cut_bins, labels = labels)

subs = (
    balance_data[balance_data['subscribed']=='yes']
    .groupby('balance_intervals', observed=False)
    .size()
)

non_subs = (
    balance_data[balance_data['subscribed']=='no']
    .groupby('balance_intervals', observed=False)
    .size()
)

percentage = (
    subs
    .div(subs + non_subs)
    .fillna(0)
    * 100
)

# Setting the data for the satisfied clients
n_satisfied_clients = balance_data[balance_data['subscribed'] == 'yes']['age'].value_counts().sort_index()

# Setting the data for the unsatisfied clients
n_unsatisfied_clients = balance_data[balance_data['subscribed'] == 'no']['age'].value_counts().sort_index()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(26, 10))

# First Graph
sns.histplot(data=balance_data, x='balance', hue='subscribed', 
             multiple='stack',          
             bins=400, 
             ax=ax1,
             palette=['red', 'green'],  
             edgecolor='.3',
             linewidth=.5)

ax1.set_ylabel("Credit")
ax1.set_xlabel("Balance")
ax1.set_title("Clients by Balance",fontsize=14)
ax1.set(xlim=(-2500, 12500))

# Second Graph
ax2.bar(labels, percentage)
ax2.set_ylabel("Percentage of Subscribers (%)")
ax2.set_xlabel("Balance Range")
ax2.set_title("Probability Subscribed by Balance",fontsize=14)
```

This analysis explores the relationship between a client’s **account balance** and the likelihood of subscribing.

The results show a clear upward trend: **the probability of subscription increases as the account balance grows**. Clients with balances above approximately **800** consistently present subscription rates higher than the dataset’s overall average subscription rate (around **7%**).

In contrast, clients with **negative balances**, indicating debt, show a significantly lower subscription rate, well below the dataset average. This suggests that customers with financial constraints are less likely to subscribe.

As balance increases further (particularly above `1500 and beyond`), the subscription rate continues to rise, reinforcing the pattern that clients with **stronger financial positions** are more likely to convert.

Overall, the analysis indicates that **account balance is a meaningful predictor of subscription behavior**, with higher balances strongly associated with a greater likelihood of subscribing.

```python
# Generating labels
label = ['jan','feb','mar','apr','may','jun','jul','aug', 'sep','oct','nov','dec']

# Setting the data for the satisfied clients
n_satisfied_clients = (
    data[data['subscribed'] == 'yes']
    .groupby('month')
    .size()
)

# Setting the data for the unsatisfied clients
n_unsatisfied_clients = (
    data[data['subscribed'] == 'no']
    .groupby('month')
    .size()
)

# Setting the index for all clients
all_clients = label

# Filling gaps of non-existing data
n_satisfied_clients = n_satisfied_clients.reindex(all_clients, fill_value=0)
n_unsatisfied_clients = n_unsatisfied_clients.reindex(all_clients, fill_value=0)

total_customers = n_satisfied_clients + n_unsatisfied_clients

percentage = (
    n_satisfied_clients
    .div(total_customers)
    .fillna(0)
    * 100
)

# Standardizing the x axis for both graphs
x_axis = label

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))

# First Graph
ax1.bar(x_axis, n_satisfied_clients, label='Yes', color='g')
ax1.bar(x_axis, n_unsatisfied_clients, bottom=n_satisfied_clients, label='No', color='r')

ax1.set_ylabel("Credit")
ax1.set_xlabel("Month")
ax1.set_title("Clients by Month",fontsize=14)
ax1.legend(loc='upper right', title='subscribed')

# Second Graph
ax2.bar(x_axis, percentage)
ax2.set_ylabel("Percentage of Subscribers (%)")
ax2.set_xlabel("Month")
ax2.set_title("Probability Subscribed by Month",fontsize=14)
```

This analysis examines whether the **month of contact** influences the probability of a client subscribing.

Some months, particularly `March` and `October`, show noticeably higher subscription rates. However, these months also have **relatively low numbers of clients**, which means the high percentages may be influenced by small sample sizes and should be interpreted cautiously.

`February` and `April` also present slightly higher subscription probabilities, though they similarly contain fewer records compared to the most common contact months.

The months with the **largest number of clients**, such as **`May`, `June`, and `July`**, show subscription rates around **6–7%**, which is close to the overall dataset average. This suggests that, in the months where most contacts occur, the probability of subscription remains relatively stable.

Overall, although some months show higher subscription rates, much of this variation can be attributed to **differences in sample size**. This suggests that the **month of contact has a strong predictive influence** on the label, though the result should be interpreted with caution.

A noticeable increase begins in the `401–600 seconds` interval, where the subscription rate reaches approximately **16%**. The rate then increases sharply in the `601–800 seconds` range, reaching around **35%**. For even longer calls, the probability continues to grow, exceeding **50%** and reaching roughly **60%** for calls longer than `1500 seconds`.

Although the number of calls decreases as duration increases, the consistent rise in subscription probability across longer call intervals indicates that **duration is one of the strongest predictors of subscription** in the dataset.

## Feature Engineering

```python
# Selecting all features that will be used
selected_features = data[['age','job','marital','education','default','balance','housing','loan','contact','duration','campaign','month','day','subscribed']]

# Selecting the categorical features
categorical_features = selected_features[['job','marital','education','contact','month','day']]

# Selecting the binary features
binary_categorical = selected_features[['default','housing','loan','subscribed']]
```

### Processing Binary Categorical Data

```python
# Transforming binary features into ones and zeros
for col in binary_categorical:
    selected_features[col] = selected_features[col].str.lower().map({'yes': 1, 'no': 0})
```

### Processing Categorical Data

```python
# Using the "one hot encoder" method to transform categorical features 
ohe = OneHotEncoder(sparse_output=False).set_output(transform='pandas')

ohe_transformation = ohe.fit_transform(categorical_features)

# Appending the new columns (and dropping the old ones)
selected_features = pd.concat([selected_features, ohe_transformation], axis=1).drop(columns=categorical_features.columns)
```

## Splitting the data into training and Testing Sets

```python
# Splitting features and label
X = selected_features.drop('subscribed', axis=1)

y = selected_features['subscribed']

# Splitting the train and test subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42 )
```

## Normalizing

```python
# Scaling features
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Defining and Training the Model

### Defining Weight

```python
# Getting number of subscribers and not subscribers
subs = (selected_features['subscribed'] == 1).sum()
non_subs = (selected_features['subscribed'] == 0).sum()

# Adjusting class weight
weight = non_subs/subs
```

Adjusting Class Weight

This step calculates the imbalance between the positive and negative classes in the dataset.

The variable `weight` is then calculated as the ratio between the majority and minority classes. This value represents how many times more frequent the negative class is compared to the positive class.

The parameter `scale_pos_weight` applies the weight on the positive class.

### Defining the Model

```python
model = XGBClassifier(scale_pos_weight=weight, random_state=42)
```

The XGBClassifier model generally handles imbalanced datasets more effectively than Random Forest. This happens because XGBoost allows the model to directly adjust the importance of the minority class through parameters such as `scale_pos_weight`, which increases the penalty for misclassifying minority samples.

Additionally, XGBoost uses gradient boosting, where each new tree is built to correct the errors of the previous ones. In imbalanced datasets, misclassified minority samples generate larger gradients, causing the model to focus more strongly on them during training.

In contrast, Random Forest builds trees independently using random sampling of the data. Although class weighting can be applied, the algorithm does not inherently focus on correcting previous classification mistakes, which can make it less sensitive to minority classes in highly imbalanced datasets.

### Training the Model

```python
model.fit(X_train_scaled, y_train)
```

## Evaluating the Model

The overall training accuracy is **93.95%**, indicating that the model correctly classified the majority of observations. However, accuracy alone is not a reliable metric in this case because the dataset is **class-imbalanced**, with many more non-subscription cases (`y = 0`) than subscription cases (`y = 1`).

Looking at the class-level metrics:

  - Precision: 1.00  
  - Recall: 0.94

  The model performs very well at identifying customers who did not subscribe.

- **Class 1 (subscribers)**

Because of the class imbalance, the **Recall for the positive class** is a more informative metric than accuracy. It shows the model's ability to correctly identify potential subscribers.

```python
display = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_training, display_labels=['0', '1'])
plt.figure(figsize=(8, 6))
display.plot(cmap=plt.cm.Reds)
plt.title('Confusion Matrix Training Set')
plt.show()
```


|                | Predicted: No Subscription (0) | Predicted: Subscription (1) |
|----------------|--------------------------------|------------------------------|

| **Actual 1**   | 7 (False Negatives)            | 2,172 (True Positives)       |

Key observations:

Given the class imbalance in the dataset, this trade-off can be acceptable depending on the application. In a marketing context, predicting more potential subscribers (even with some false positives) may be preferable to missing actual customers who would subscribe.

#### Precision Recall Display

The Precision–Recall curve generated using cross-validation shows how the classifier balances **precision** and **recall** across different decision thresholds. The model achieves an **AUPRC of approximately 0.51**, which indicates a moderate ability to distinguish the positive class from the negative class.

At **low recall levels**, the model maintains relatively **high precision (around 0.65–0.70)**, meaning that when the model is conservative in predicting the positive class, its predictions are more reliable. However, as recall increases, **precision gradually decreases**, eventually dropping below **0.3** when recall approaches 1. This behavior indicates that capturing more positive samples leads to a larger number of false positives.

Since this curve is computed using **cross-validated predictions**, it provides a more reliable estimate of the model's generalization performance compared to evaluating on a single training split.

Overall, the model demonstrates a **reasonable trade-off between precision and recall**.
